## Team  Statistics 

```python
#- Dropping rows with no result 
no_res_index = df_ana[df_ana.result=='no result'].index
df_ana.drop(no_res_index,axis=0,inplace=True)
```

```python
# Win Count
df_win = df_ana.winner.value_counts()

# Toss Won Count
df_toss_win = df_ana.toss_winner.value_counts()

# MWTW - Match Won Toss Won
df_win_mt = df_ana[df_ana.toss_winner == df_ana.winner].winner.value_counts()

# Number of Matches Played 
df_match_cnt = pd.DataFrame([df_ana.team1.value_counts(),df_ana.team2.value_counts()]).transpose()
df_match_cnt['Total_Matches'] = df_match_cnt.team1 + df_match_cnt.team2
df_stats = pd.DataFrame([df_match_cnt['Total_Matches'],df_win,df_win_mt,df_toss_win]).transpose()
df_stats.columns = ['Total_Matches','Matches_Won','MWTW','Toss_Won']

# Matches Lost = Matches played - Matches Won 
df_stats['Matches_Lost'] = df_stats.Total_Matches - df_stats.Matches_Won

# MWTL - Match Won Toss Lost
# Matches Won when toss lost = Matches Won - Matches won when toss won 
df_stats['MWTL'] = df_stats.Matches_Won - df_stats.MWTW

# Toss Lost Count
df_stats['Toss_Lost'] = df_stats.Total_Matches - df_stats.Toss_Won

# MLTW - Match Lost Toss Won
# Toss Won - MWTW
df_stats['MLTW'] = df_stats.Toss_Won - df_stats.MWTW

# MLTL - Match Lost Toss Lost
# Matches Lost - MLTW
df_stats['MLTL'] = df_stats.Matches_Lost - df_stats.MLTW

df_stats[['Total_Matches','Matches_Won','Matches_Lost',
          'Toss_Won','Toss_Lost','MWTW','MWTL','MLTW','MLTL']].head(2)

#- Win Percentage
df_stats['Win_Percentage'] =  (df_stats.Matches_Won / df_stats.Total_Matches) * 100 

#- Lost Percentage
df_stats['Lost_Percentage'] = (df_stats.Matches_Lost / df_stats.Total_Matches) * 100
```

### Win-Loss Summary 

```python
df_stats[['Total_Matches','Matches_Won','Matches_Lost']].plot(kind='bar',
                                                              figsize=(12,8),
                                                              fontsize=13)
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.title('Teams Win-Loss Stats ')
plt.savefig('teams_win_loss_stats.png')  
```

### Toss Decision Bat/Field Impact on Match 

```python

df_field = df_ana[(df_ana.toss_winner == df_ana.winner) & ( df_ana.toss_decision == "field")][['toss_winner','toss_decision']]
df_bat = df_ana[(df_ana.toss_winner == df_ana.winner) & ( df_ana.toss_decision == "bat")][['toss_winner','toss_decision']]

toss_dec_stats = pd.DataFrame(columns=['field','bat','total'])
toss_dec_stats['total'] = df_stats.Total_Matches
toss_dec_stats['field'] = (df_field.toss_winner.value_counts()/toss_dec_stats.total) * 100 
toss_dec_stats['bat'] = (df_bat.toss_winner.value_counts()/toss_dec_stats.total ) * 100

toss_dec_stats[['field','bat']].sort_values(['field','bat'],ascending=False).plot(kind='bar',
                                     figsize=(10,8),
                                     fontsize=12)
plt.xlabel('Teams')
plt.ylabel('Matches Win Percentage')
plt.title('Toss Decision Impact on Matches')
plt.savefig('toss_decision_impact_on_matches.png')  

                 
print("Max Percentage Win When Choose to Field = {}".format(toss_dec_stats.field.max()))
print("Max Percentage Win When Choose to Bat = {}".format(toss_dec_stats.bat.max()))     
```

Observations:
    
    1. Gujarat Lions Won 33.33% of their Matches when they won the Toss and Choose to Field
    2. Chennai Super Kings Won 19.72% of their Matches when they won the Toss and Choose to bat
